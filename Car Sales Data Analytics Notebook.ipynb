```python
#Getting yearly total sales
yearly_sales = data.groupby('Car Year')['Sale Price'].sum().reset_index()
yearly_sales = yearly_sales.sort_values(by='Sale Price', ascending=False)
print(yearly_sales)
```

```python
import matplotlib.pyplot as plt

# Extraction of year and sales data
years = yearly_sales['Car Year']
sales = yearly_sales['Sale Price']

# Calculation of total sales (assuming all sales data is present)
total_sales = sales.sum()

# Calculate sales percentages
sales_percentages = (sales / total_sales) * 100

# Creation of the line chart
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.plot(years, sales_percentages, marker='o', linestyle='-')

# Setting labels and title
plt.xlabel('Car Year')
plt.ylabel('Sales Percentage (%)')
plt.title('Yearly Sales Percentage')

# Customization of the plot 
plt.grid(True)  # grid lines for better readability
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility with many years

# Displaying the chart
plt.tight_layout()  
plt.show()
```

* The year 2015 having the highest sales and the year 2022 having the lowest sales.
